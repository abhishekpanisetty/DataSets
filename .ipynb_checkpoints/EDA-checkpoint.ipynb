```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from matplotlib import cm
import seaborn as sns
```

```python
cab=pd.read_csv("Cab_Data.csv")
```

```python
city=pd.read_csv("City.csv")
city['Population'] = city['Population'].str.replace(',','')
city['Users'] = city['Users'].str.replace(',','')
city['Population']=city['Population'].astype(int)
city['Users']=city['Users'].astype(int)
```

```python
customer=pd.read_csv("Customer_ID.csv")
```

```python
transaction=pd.read_csv("Transaction_ID.csv")
```

```python
city
```

```python
cab
```

```python
customer
```

```python
transaction
```

```python
#Merging customer data and transaction data based on Customer_ID
cus_tran=pd.DataFrame.merge(customer,transaction,how="outer",on="Customer ID")
cus_tran
```

```python
#Calulating profit of each transaction by using Price_Charged - Cost of trip
cab["profit"]=cab["Price Charged"]-cab["Cost of Trip"]
cab
```

```python
#pie chart to represent percentage of users in each city
colors = sns.color_palette('bright')[0:5]
plt.figure(figsize=(15,15))
plt.pie(city["Users"], labels = city["City"], colors = colors, autopct='%.0f%%')
plt.show()
```

```python
#Merging all the data frames to one data frame
final=pd.DataFrame.merge(cab,cus_tran,on="Transaction ID")
final
```

```python
#Finding the sum of total number of KMS travelled by each cab company
Total_km_travelled=final.groupby('Company')['KM Travelled'].sum()
```

```python
Total_km_travelled
```

```python
#Pie chart for total number of KMS travelled by each cab company
plt.pie(Total_km_travelled, labels=Total_km_travelled.index, autopct='%.0f%%')
```

```python
#Finding the total sum of Price Charged by each cab company for all the transactions
Total_price_charged=final.groupby('Company')['Price Charged'].sum()
```

```python
#finding the unique dates in the dataset
unique_values = final['Date of Travel'].unique()
unique_values
```

```python
#converting the series to dataframe
uniques = pd.DataFrame(data=unique_values)
```

```python
uniques.columns=["Date of Travel"]
uniques=uniques.sort_values("Date of Travel")
```

```python
#converting it into proper date format
uniques['date'] = pd.date_range(start='31/01/2016', periods=len(uniques), freq='D')
```

```python
uniques
```

```python
#Merging the 2 data frames based on a column
final=pd.DataFrame.merge(final,uniques,on="Date of Travel")
```

```python
final
```

```python
#Splitting year and month
final['year'], final['month'] = final['date'].apply(lambda x: x.year), final['date'].apply(lambda x: x.month)
```

```python
final
```

```python
fig,axes = plt.subplots(1, 2, figsize=(25, 8), sharey=True)
fig.suptitle('Boxplot Distributions')
sns.scatterplot(ax=axes[0],data=final, x='KM Travelled', y='Price Charged' , hue="Company",palette="husl"  ).set_title("KM_Travelled  - Price_Charged")
sns.scatterplot(ax=axes[1],data=final, x='KM Travelled', y='Cost of Trip' , hue="Company",palette="husl").set_title("KM_Travelled  - Cost_of_Trip")
```

```python
fig,axes = plt.subplots(figsize=(16, 6), sharey=True)
fig.suptitle('Boxplot Distributions')
sns.scatterplot(data=final, x='Price Charged', y='Cost of Trip' , hue="Company").set_title("Price_Charged - Cost_of_Trip")
```

```python
fig,axes = plt.subplots(figsize=(16, 6), sharey=True)
fig.suptitle('Boxplot Distributions')
sns.scatterplot(data=final, x="KM Travelled", y='profit' , hue="Company").set_title("KM Travelled - Profit")
```

```python
final.corr(method ='pearson')
```

```python
plt.figure(figsize=(16, 6))
# Store heatmap object in a variable to easily access it when you want to include more features (such as title).
# Set the range of values to be displayed on the colormap from -1 to 1, and set the annotation to True to display the correlation values on the heatmap.
heatmap = sns.heatmap(final.corr(), vmin=-1, vmax=1, annot=True)
# Give a title to the heatmap. Pad defines the distance of the title from the top of the heatmap.
heatmap.set_title('Correlation Heatmap', fontdict={'fontsize':12}, pad=12)
```

```python
#Splitting the data into individual cab data
PinkCab = final[final['Company']=='Pink Cab']
YellowCab = final[final['Company']=='Yellow Cab']
```

```python
#Figuring the pie chart for Yellow Cab Company for all the years
yellow_profit_per_year = YellowCab.groupby("year")["profit"].sum()
prof_fig = px.pie(profit_per_year,values=yellow_profit_per_year,names=yellow_profit_per_year.index ,title="Profit per year for yellow cab company")
prof_fig.update_traces(textposition='inside',textinfo='percent+label')
prof_fig.update_layout( width = 1100 , height = 600 )
```

```python
#Figuring the pie chart for Pink Cab Company for all the years
pink_profit_per_year = PinkCab.groupby("year")["profit"].sum()
prof_fig = px.pie(profit_per_year,values=pink_profit_per_year,names=pink_profit_per_year.index ,title="Profit per year for Pink cab company")
prof_fig.update_traces(textposition='inside',textinfo='percent+label')
prof_fig.update_layout( width = 1100 , height = 600 )
```

```python
#Plotting the profits for each cab company over the years
import plotly.graph_objects as go
fig = go.Figure(data=[go.Bar(name='Yellow Cab', x=yellow_profit_per_year.index, y=yellow_profit_per_year),
                      go.Bar(name='Pink Cab', x=pink_profit_per_year.index, y=pink_profit_per_year)])
fig.update_layout(barmode='group')
fig.show()
```
